8. ARPU и p-value

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_excel('Данные АБ тестов.xlsx')
df.head()
```

```python
group_sizes = df.groupby(['experiment_num', 'experiment_group']).size().reset_index(name='user_count')
print("Количество пользователей по группам:\n")
print(group_sizes)
```

```python
for i in [1, 2, 3]:
    print(f"\nЭксперимент {i}")
    exp_data = df[df['experiment_num'] == i]
    control_rev = exp_data[exp_data['experiment_group'] == 'control']['revenue']
    test_rev = exp_data[exp_data['experiment_group'] == 'test']['revenue']
    
    _, p_shapiro_ctrl = stats.shapiro(control_rev)
    _, p_shapiro_test = stats.shapiro(test_rev)
    
    print(f"[Шапиро-Уилк] p-value Контроль: {p_shapiro_ctrl:.6f}")
    print(f"[Шапиро-Уилк] p-value Тест:     {p_shapiro_test:.6f}")
    
    if p_shapiro_ctrl < 0.05 or p_shapiro_test < 0.05:
        print("Результат: Распределения не нормальные.")
    else:
        print("Результат: Распределения нормальные.")
        
    stat_mw, p_value_mw = stats.mannwhitneyu(test_rev, control_rev, alternative='two-sided')
    print(f"[Манн-Уитни]  p-value:          {p_value_mw:.6f}")
    
    if p_value_mw < 0.05:
        print("Результат: Различия в распределениях статистически значимы.")
    else:
        print("Результат: Различия незначимы.")
```
